```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/german_credit_data.csv")

df
```

```python
print("### Statistik Deskriptif Kolom Numerik ###")
print(df.describe())
```

# 1. DATA CLEANSING

```python
# a. Hapus kolom tak penting (Unnamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

df
```

```python
# b. Cek dan hapus duplikasi
print("Jumlah duplikat:", df.duplicated().sum())
df = df.drop_duplicates()

df
```

```python
# c. Menghitung jumlah missing values
df[['Duration']].isnull().sum()


# Mengisi missing dengan nilai 0
# yang mengindikasikan proses belum selesai, sehingga durasinya dianggap 0 untuk saat ini.
df['Duration'] = df['Duration'].fillna(0)
df
```

```python
#d. menangani outlier

# Hitung Q1, Q3, dan IQR
Q1          = df['Credit amount'].quantile(0.25)
Q3          = df['Credit amount'].quantile(0.75)
IQR         = Q3 - Q1
batas_atas  = Q3 + 1.5 * IQR
batas_bawah = Q1 - 1.5 * IQR

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Batas Atas: {batas_atas}, Batas Bawah: {batas_bawah}")

# Deteksi outlier
outliers = df[(df['Credit amount'] > batas_atas) | (df['Credit amount'] < batas_bawah)]
outliers

# Capping outlier pada kolom 'jumlah_pinjaman'
df['Credit amount'] = df['Credit amount'].clip(
   lower = batas_bawah,
   upper = batas_atas
)

df
```

# 2. Transformasi Data dan Rekayasa Fitur Sederhana

```python
# a. Membuat fitur 'kategori_Age' dengan binning
# 'umur' menjadi 3 kategori: Muda, Dewasa, Lansia
bins_Age                      = [0, 30, 60, np.inf]
labels_Age                    = ['Young', 'Adult', 'Elderly']

df['Age_category'] = pd.cut(df['Age'], bins=bins_Age, labels=labels_Age)

df
```

```python
# b1. membuat fitur 'Duration_month'
# Kita gunakan kolom 'Duration' yang berisi hari, kemudian dibagi 30
# dan isi NaN dengan 0.

df['Duration_month'] = (df['Duration'].fillna(0) / 30)

df
```

```python
# b2. membuat fitur 'Duration_category'

df['Duration_category'] = pd.cut(
    df['Duration_month'],
    bins=[-1, 1, 2, df['Duration_month'].max()],
    labels=['Short', 'Medium', 'Long']
)

df
```

```python
# c. Terapkan Ordinal Encoding pada kolom 'Saving Account'

mapping_Saving_accounts = {'little': 1, 'moderate': 2, 'quite rich': 3, 'rich': 4}

df['Saving_accounts_oe'] = df['Saving accounts'].map(mapping_Saving_accounts)
df
```

MULTIVARIATE ANALYSIS
