## **2. CHECK DATA QUALITY & SUMMARY STATISTIC**

```python
#Tampilkan tipe data (dtypes) dan jumlah nilai kosong (null values) di setiap kolom.
print("### Ringkasan Teknis (df_credit_data.info()) ###")
df_credit_data.info()
```

```python
#Tampilkan statistik deskriptif untuk: Kolom numerik dan Kolom kategorikal

print("### Statistik Deskriptif Kolom Numerik ###")
print(df_credit_data.describe())

print("\n### Statistik Deskriptif Kolom Kategorikal ###")
print(df_credit_data.describe(include=['object']))
```

```python
#Identifikasi jumlah missing value di setiap kolom

print("### Jumlah Missing Value per Kolom ###")
print(df_credit_data.isnull().sum())
```

## **3. UNIVARIATE ANALYSIS**

```python
#Pilih minimal 3 variabel untuk dianalisis distribusinya.
#Tuliskan insight singkat dari masing-masing analisis

# Menampilkan jumlah Sex
print("### Distribusi Kolom 'Sex' ###")
print(df_credit_data ['Sex'].value_counts())
```

Berdasarkan jenis kelamin mayoritas pengguna kredit adalah  pria (69%) dibandingkan wanita (31%), yang menunjukkan bahwa pria lebih dominan dalam memanfaatkan fasilitas kredit meskipun wanita juga memiliki peran signifikan.

```python

# Menampilkan Age
print("### Statistik Deskriptif Kolom 'Age' ###")
print(df_credit_data ['Age'].describe())
```

Mayoritas pengguna kredit berasal dari usia produktif, yaitu 27–42 tahun,
yang menunjukkan bahwa fasilitas kredit paling banyak dimanfaatkan oleh individu yang sedang berada pada fase aktif bekerja.


```python
# menampilkan housing
print("### Distribusi Kolom 'Housing' ###")
print(df_credit_data ['Housing'].value_counts())
```


Berdasarkan distribusi kolom Housing, mayoritas pengguna kredit memiliki rumah sendiri, diikuti dengan penyewa rumah, dan sebagian kecil tinggal gratis, yang menunjukkan bahwa fasilitas kredit lebih banyak dimanfaatkan oleh individu dengan kepemilikan rumah tetap.


## **4. MULTIVARIATE ANALYSIS**

```python
#Pilih minimal 2 jenis analisis hubungan antar variabel seperti:
#Filtering data
#Group by & aggregation
#Cross tabulation
#Korelasi (heatmap)

# Pastikan metode yang digunakan relevan untuk mencari pola awal dari data.

# Crosstab antara Sex dan Risk
crosstab_sex_risk = pd.crosstab(df_credit_data['Sex'], df_credit_data['Risk'])
print("### Cross Tabulation: Sex vs Risk ###")
print(crosstab_sex_risk)
```

```python
# Group by Purpose
purpose_credit = df_credit_data.groupby('Purpose')['Credit amount'].mean().sort_values(ascending=False)
print("### Rata-rata Credit Amount berdasarkan Purpose ###")
print(purpose_credit)
```

## **5. CONCLUSION**

*   Berdasarkan crosstab jenis kelamin dan Risiko, Mayoritas pria maupun wanita tergolong good risk, tetapi pria memiliki proporsi good risk yang lebih tinggi dibanding wanita , sehingga dapat disimpulkan bahwa secara umum, pria dianggap lebih layak kredit dibanding wanita.
*   Berdasarkan rata-rata jumlah kredit berdasarkan Purpose, tujuan dengan kredit terbesar adalah vacation/others , jauh lebih tinggi dibanding tujuan lain, sedangkan yang terendah adalah domestic appliances, sehingga dapat disimpulkan bahwa pinjaman dengan nominal besar umumnya digunakan untuk kebutuhan liburan/tujuan lain, sementara pembelian peralatan rumah tangga cenderung membutuhkan dana lebih kecil.
