# EXERCISE DATA MANIPULATION

##TASK
**1.   Manipulasi Data Waktu (Datetime)**
*   Konversi kolom tanggal_aplikasi dan tanggal_pembayaran menjadi tipe data datetime.
*   Buat kolom baru bernama durasi_pinjaman_hari yang menghitung selisih antara tanggal_pembayaran dan tanggal_aplikasi dalam satuan hari.


**2.   Manipulasi Data Teks**
* Menghilangkan spasi ekstra di awal dan akhir string (.str.strip())
* Mengubah semua huruf menjadi huruf kecil (.str.lower()) untuk menstandarisasi data

**3.  Penanganan Missing Value**
* Temukan missing value di kolom jumlah_pembayaran
* Isi missing value tersebut dengan nilai median dari kolom jumlah_pembayaran yang sudah ada

**4. Analisis dan Agregasi**
* Gunakan .groupby() untuk menghitung rata-rata durasi_pinjaman_hari dan total jumlah_pembayaran untuk setiap status_pinjaman
* Sajikan hasil dalam format tabel yang mudah dibaca


**Data Frame**

```python
import pandas as pd
import numpy as np

# Data Aplikasi Pinjaman
data_aplikasi = {
    'id_pinjaman': [101, 102, 103, 104, 105],
    'id_pelanggan': [1, 2, 3, 4, 5],
    'tanggal_aplikasi': ['2023-01-10', '2023-01-25', '2023-02-05', '2023-02-15', '2023-03-01'],
    'status_pinjaman': ['  lunas', 'Gagal Bayar', 'Lunas', '  lunas', 'Gagal Bayar ']
}
df_aplikasi = pd.DataFrame(data_aplikasi)

df_aplikasi
```

```python
# Data Pelanggan
data_pelanggan = {
    'id_pelanggan': [1, 2, 3, 4, 5],
    'nama': ['Budi', ' anita', 'Charlie', 'David', 'Erika'],
    'kota_asal': ['jakarta', ' Bandung ', 'surabaya ', 'jakarta', 'medan']
}
df_pelanggan = pd.DataFrame(data_pelanggan)

df_pelanggan
```

```python
# Data Histori Pembayaran (terdapat missing value)
data_pembayaran = {
    'id_pinjaman': [101, 102, 103, 104, 105],
    'tanggal_pembayaran': ['2023-03-10', '2023-03-20', '2023-03-05', '2023-04-10', '2023-05-01'],
    'jumlah_pembayaran': [15000000, 1000000, 7500000, np.nan, 2000000]
}
df_pembayaran = pd.DataFrame(data_pembayaran)

df_pembayaran
```

**1. Manipulasi Data Waktu (Datetime)**

a. Konversi kolom tanggal_aplikasi dan tanggal_pembayaran menjadi tipe data datetime.


```python
# Konversi kolom 'tanggal_aplikasi'
df_aplikasi['tanggal_aplikasi_terformat'] = pd.to_datetime(df_aplikasi['tanggal_aplikasi'])
df_aplikasi
```

```python
#Konversi kolom 'tanggal_pembayaran'
df_pembayaran['tanggal_pembayaran_terformat'] = pd.to_datetime(df_pembayaran['tanggal_pembayaran'])
df_pembayaran
```

b. Buat kolom baru bernama durasi_pinjaman_hari yang menghitung selisih antara tanggal_pembayaran dan tanggal_aplikasi dalam satuan hari.

```python
#Membuat kolom durasi_pinjaman_hari
df_Aplikasi_Pembayaran = pd.merge(df_aplikasi, df_pembayaran, on='id_pinjaman', how='outer')
df_Aplikasi_Pembayaran

```

```python
# Perhitungan durasi

# Kolom baru durasi_pinjaman_hari
df_Aplikasi_Pembayaran['durasi_pinjaman_hari'] = (
    df_Aplikasi_Pembayaran['tanggal_pembayaran_terformat'] - df_Aplikasi_Pembayaran['tanggal_aplikasi_terformat']
).dt.days

df_Aplikasi_Pembayaran
```

**2.Manipulasi Data Teks**

Bersihkan kolom status_pinjaman dan kota_asal dengan

a. Menghilangkan spasi ekstra di awal dan akhir string (.str.strip())

```python
#Gabungkan Tabel data aplikasi, data pelanggan, dan data pembayaran
df_Aplikasi_Pembayaran_Pelanggan = pd.merge(df_Aplikasi_Pembayaran, df_pelanggan, on='id_pelanggan', how='outer')
df_Aplikasi_Pembayaran_Pelanggan

#Menghilangkan spasi ekstra di awal dan akhir string
df_Aplikasi_Pembayaran_Pelanggan['status_pinjaman_strip'] = df_Aplikasi_Pembayaran_Pelanggan['status_pinjaman'].str.strip()
df_Aplikasi_Pembayaran_Pelanggan['kota_asal_strip'] = df_Aplikasi_Pembayaran_Pelanggan['kota_asal'].str.strip()

df_Aplikasi_Pembayaran_Pelanggan

```

b. Mengubah semua huruf menjadi huruf kecil (.str.lower()) untuk menstandarisasi data.

```python
#Mengubah semua huruf menjadi huruf kecil
df_Aplikasi_Pembayaran_Pelanggan['status_pinjaman_lower'] = df_Aplikasi_Pembayaran_Pelanggan['status_pinjaman_strip'].str.lower()
df_Aplikasi_Pembayaran_Pelanggan['kota_asal_lower'] = df_Aplikasi_Pembayaran_Pelanggan['kota_asal_strip'].str.lower()

df_Aplikasi_Pembayaran_Pelanggan
```

**3. Penanganan Missing Value**

a. Temukan missing value di kolom jumlah_pembayaran.

```python
#Menghitung jumlah null

df_Aplikasi_Pembayaran_Pelanggan['jumlah_pembayaran'].isnull().sum()
```

b. Isi missing value tersebut dengan nilai median dari kolom jumlah_pembayaran yang sudah ada.

```python
#Menghisi missing value dengan median

df_Aplikasi_Pembayaran_Pelanggan['jumlah_pembayaran'] = df_Aplikasi_Pembayaran_Pelanggan['jumlah_pembayaran'].fillna(
    df_Aplikasi_Pembayaran_Pelanggan['jumlah_pembayaran'].median())

df_Aplikasi_Pembayaran_Pelanggan
```

**4. Analisis dan Agregasi**

a. Gunakan .groupby() untuk menghitung rata-rata durasi_pinjaman_hari dan total jumlah_pembayaran untuk setiap status_pinjaman.


```python
#Menghitung rata-rata durasi_pinjaman_hari
rata_rata_per_status = (
    df_Aplikasi_Pembayaran_Pelanggan.groupby('status_pinjaman_lower')['durasi_pinjaman_hari']
    .mean()
    .reset_index(name='rata_rata_durasi_pinjaman_hari')
)

print("\nRata-rata durasi pinjaman per status pinjaman:")
print(rata_rata_per_status)

print('-'*50)

#Menghitung total jumlah_pembayaran untuk setiap status pinjaman
total_pembayaran_per_status = (
    df_Aplikasi_Pembayaran_Pelanggan.groupby('status_pinjaman_lower')['jumlah_pembayaran']
    .sum()
    .reset_index(name='total_jumlah_pembayaran')
)

print("\nTotal Jumlah Pembayaran per status pinjaman:")
print(total_pembayaran_per_status)
```

b. Sajikan hasil dalam format tabel yang mudah dibaca.

```python
# Gabungkan tabel
tabel_ringkasan = pd.merge(rata_rata_per_status, total_pembayaran_per_status, on='status_pinjaman_lower')

print("\nRingkasan Rata-rata Durasi & Total Pembayaran per Status Pinjaman:")
print(tabel_ringkasan.to_string(index=False))
```
